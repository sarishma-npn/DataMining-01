```python
#@ import dataset using pandas library
import pandas as pd
df=pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Lab 1/Customer Call List.xlsx')
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.drop(columns="Not_Useful_Column")

df
```

```python
#@ removing unnecessary character from last name
df["Last_Name"]=df["Last_Name"].str.strip("123._/")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'\D', '', regex=True)
display(df)
```

```python
#@ prase address column into Street_Address,State and Zip_code
```

```python
#@ Lets make playing customer to be Y for Yes and N for No
df['Paying Customer']=df['Paying Customer'].replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].replace('No','N')
df
```

```python
# Replace 'N/a' and NaN with empty string
df = df.replace(['N/a', pd.NA, None], '')
df = df.fillna('')
display(df)
```

```python
#@ Lets make playing customer to be Y for Yes and N for No
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].replace('No','N')
df
```

```python
df = df[df['Do_Not_Contact'] != 'Y']
display(df)
```

```python
#@ Drop the row if phone number is blank
df = df[df['Phone_Number'] != ''].copy()
display(df)
```

```python
#@ arrange the order of the data 1,2,3
df = df.reset_index(drop=True)
display(df)
```
